```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("nonStrategicAnalysis.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["5", "10", "15", "20", "25", "30", "35", "40","45"]


df['45'] = df['45'].str.strip(']').astype(float)
df['5'] = df['5'].str.strip('[').astype(float)


mean = df.mean()
stds = df.std()
print("Means of each column: ")
print(mean)
print("Sd of each columns: ")
print(stds)

plt.figure(figsize=(10, 6))
plt.scatter(mean.index, mean.values, color='blue')

# Add title and labels
plt.title('Mean of Each Column')
plt.xlabel('Columns')
plt.ylabel('Mean Value')

# Show plot
plt.grid(True)
plt.show()
```
