```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("notConstantVar0.8.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "alpha", "numCustomer", "meanNoRe", "SDNoRe", "meanWithRe", "SDRe", "meanAllQueue", "SDAllQueue", "meanAllLine", "SDAllLine"]


df['SDAllLine'] = df['SDAllLine'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])


for i in range(1, 5):
    df_plot = df[df['onlineBarista'] == i]
    df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
   
    print("df_plot dataframe is below")
    print(df_plot)
       
    plt.plot(df_plot['numCustomer'], df_plot['meanWithRe'], label = "GreedyWithReallocation")
    plt.fill_between(df_plot['numCustomer'], df_plot['meanWithRe'] -  df_plot['SDRe'], df_plot['meanWithRe'] +  df_plot['SDRe'], color='b', alpha=0.1)
   
    plt.plot(df_plot['numCustomer'], df_plot['meanNoRe'], label = "GreedyWithoutReallocation")
    plt.fill_between(df_plot['numCustomer'], df_plot['meanNoRe'] -  df_plot['SDNoRe'], df_plot['meanNoRe'] +  df_plot['SDNoRe'], color='r', alpha=0.1)

    plt.plot(df_plot['numCustomer'], df_plot['meanAllQueue'], label = "AllQueue")
    plt.fill_between(df_plot['numCustomer'], df_plot['meanAllQueue'] -  df_plot['SDAllQueue'], df_plot['meanAllQueue'] +  df_plot['SDAllQueue'], color='g', alpha=0.1)

    plt.plot(df_plot['numCustomer'], df_plot['meanAllLine'], label = "AllLine")
    plt.fill_between(df_plot['numCustomer'], df_plot['meanAllLine'] -  df_plot['SDAllLine'], df_plot['meanAllLine'] +  df_plot['SDAllLine'], color='y', alpha=0.1)
   
    plt.title("Num of online baristas: " + str(i) + ", Num of physical baristas:" + str(5-i))
    plt.xlabel("Number of customers")
    plt.ylabel("Total cost")
    plt.legend(loc="upper left", fontsize = "5")

    plt.savefig("C:\\Users\\Kiet Nguyen\\Documents\\UofR\\SUMMER_2024\\EmpiricalResults\\NotConstatVariable\\Alpha=0.8\\" + str(i) + "online" + str(5-i) + "physical" + ".pdf", bbox_inches='tight' )

        #plt.title("Online barista: " + str(i) + ", Physical barista: " + str(5-i))
    plt.show()

        
       

print(df.dtypes)

```
