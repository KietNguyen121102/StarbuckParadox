```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

df = pd.read_csv("semiStrategicComparison.txt", sep = ",")
df = pd.DataFrame(df)
df.columns = ["onlineBarista","physicalBarista", "physicalOverOnline", "greedyPopulation" , "Re", "NoRe"]


df['NoRe'] = df['NoRe'].str.strip(']').astype(float)
df['onlineBarista'] = df['onlineBarista'].str.strip('[').astype(float)


pd.to_numeric(df['onlineBarista'])
pd.to_numeric(df['physicalBarista'])
pd.to_numeric(df['physicalOverOnline'])

# df['physicalOverOnline'].round(2)
uniques = df['physicalOverOnline'].unique()
print(uniques)

# for i in range(0,6):
for value in uniques:
        for i in range(0,6):
                df_plot = df[df['onlineBarista'] == i]
                df_plot = df_plot[df_plot['physicalBarista'] == 5-i]
        
                print(value)
                plot = df_plot[df_plot['physicalOverOnline'] == value]
                # plt.plot(plot['greedyPopulation'], plot['Re']/100, label = 'Online Barista: ' + str(i))
                plt.plot(plot['greedyPopulation'], plot['NoRe']/100, label = 'NoRe Online Barista: ' + str(i))

               
                plt.xlabel("Greedy Population")
                plt.ylabel("Average cost")
                plt.legend(loc="upper right", fontsize = "5")

                plt.savefig("C:\\Users\\Kiet Nguyen\\Documents\\UofR\\SUMMER_2024\\EmpiricalResults\\SemiStrategic\\WithoutReallocation\\" + "PhysicalRatio=" + str(value) + ".pdf", bbox_inches='tight' )
                # plt.legend(plot['physicalOverOnline'])
                if value == 0.6000000000000001:
                        plt.title("Physical/(1-Greedy): " + str(0.6))
                else:
                        plt.title("Physical/(1-Greedy): " + str(value))


        
        plt.show()
      
       
        #plt.title("Online barista: " + str(i) + ", Physical barista: " + str(5-i))



        
       

print(df.dtypes)
```
